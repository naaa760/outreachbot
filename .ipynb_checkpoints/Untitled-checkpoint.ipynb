```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://jobs.nike.com/?jobSearch=true&jsOffset=0&jsSort=posting_start_date&jsLanguage=en")
page_data = loader.load().pop().page_content
print(page_data)
```

```python
from langchain_core.prompts import PromptTemplate

# Define the scraped page data (replace this with the actual text you scraped)
page_data = """
We are hiring:
- Role: Software Engineer
  Experience: 2+ years
  Skills: Python, Django, APIs
  Description: Build and maintain web applications.

- Role: Data Scientist
  Experience: 3+ years
  Skills: Python, Machine Learning, Statistics
  Description: Analyze and interpret complex data to help decision-making.
"""

# Create the prompt template
prompt_extract = PromptTemplate.from_template(
    """
    ### SCRAPED TEXT FROM WEBSITE:
    {page_data}
    ### INSTRUCTION:
    The scraped text is from the career's page of a website.
    Your job is to extract the job postings and return them in JSON format containing the 
    following keys: `role`, `experience`, `skills` and `description`.
    Only return the valid JSON.
    ### VALID JSON (NO PREAMBLE):    
    """
)

# Assume `llm` is defined as your language model (e.g., OpenAI LLM or other)
chain_extract = prompt_extract | llm 

# Invoke the chain with the page_data
res = chain_extract.invoke(input={'page_data': page_data})

# Check the result
print(type(res.content))
print(res.content)
```

```python
from langchain_core.output_parsers import JsonOutputParser

json_parser = JsonOutputParser()
json_res = json_parser.parse(res.content)
json_res
```

```python
import pandas as pd

df = pd.read_csv("my_portfolio.csv")
df
```

```python
import uuid
import chromadb

client = chromadb.PersistentClient('vectorstore')
collection = client.get_or_create_collection(name="portfolio")

if not collection.count():
    for _, row in df.iterrows():
        collection.add(documents=row["Techstack"],
                       metadatas={"links": row["Links"]},
                       ids=[str(uuid.uuid4())])
```

```python
# Define the job variable with relevant data
job = {
    'skills': 'Python, React, SQL'  # Replace this with the actual skills for the job
}

# Now run the query
links = collection.query(query_texts=job['skills'], n_results=2).get('metadatas', [])
print(links)
```

```python
job
```

```python
prompt_email = PromptTemplate.from_template(
        """
        ### JOB DESCRIPTION:
        {job_description}
        
        ### INSTRUCTION:
        You are Mohan, a business development executive at AtliQ. AtliQ is an AI & Software Consulting company dedicated to facilitating
        the seamless integration of business processes through automated tools. 
        Over our experience, we have empowered numerous enterprises with tailored solutions, fostering scalability, 
        process optimization, cost reduction, and heightened overall efficiency. 
        Your job is to write a cold email to the client regarding the job mentioned above describing the capability of AtliQ 
        in fulfilling their needs.
        Also add the most relevant ones from the following links to showcase Atliq's portfolio: {link_list}
        Remember you are Mohan, BDE at AtliQ. 
        Do not provide a preamble.
        ### EMAIL (NO PREAMBLE):
        
        """
        )

chain_email = prompt_email | llm
res = chain_email.invoke({"job_description": str(job), "link_list": links})
print(res.content)
```
